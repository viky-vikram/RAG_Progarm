# 2. RAG Types

There are different **types of RAG**. Each one improves the basic RAG in a different way to give **better and more accurate** answers.

This notebook gives a **quick overview**. We will learn each type in detail in the next notebooks.

## Quick recap: Basic (Naive) RAG

The simplest RAG does just 3 steps:

```
Retrieve  ->  Augment  ->  Generate
```

It works, but sometimes it:
- Picks **wrong or weak** documents.
- Cannot **check** if its answer is correct.

The advanced types below fix these problems.

## The main types of RAG

| Type | Main idea (in one line) |
|------|--------------------------|
| **Self RAG** | The AI checks and corrects **itself** while answering. |
| **Corrective RAG (CRAG)** | If retrieved data is **bad**, it searches again (e.g. the web). |
| **Fusion RAG** | Asks the question in **many ways** and **combines** the results. |
| **BM25 + Rerank** | Uses keyword search + **re-orders** results by relevance. |

## Simple way to remember each type

- **Self RAG** → "Let me double-check my own answer."
- **Corrective RAG** → "This info is bad, let me search again."
- **Fusion RAG** → "Let me ask the question in many ways."
- **BM25 Rerank** → "Find by keywords, then sort the best on top."

## Key points to remember

- Basic RAG = Retrieve ➜ Augment ➜ Generate.
- Advanced RAG types make answers **more accurate**.
- **Self RAG** → self-checking.
- **Corrective RAG** → re-search when data is bad.
- **Fusion RAG** → combine many search results.
- **BM25 Rerank** → keyword search + re-ordering.
