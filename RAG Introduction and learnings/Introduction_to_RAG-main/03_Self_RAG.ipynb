# 3. Self RAG

**Self RAG** (Self-Reflective RAG) is a type of RAG where the AI **checks its own work**.

Instead of blindly using whatever it finds, the AI keeps asking itself:
- *Do I even need to search for this?*
- *Is the information I found useful?*
- *Does my answer actually match the information?*

## Simple real-life example

Think of a careful student writing an answer:

1. "Do I need to look this up, or do I already know it?"
2. "Is this book page actually relevant?"
3. "Did I answer correctly based on the page?"

Self RAG makes the AI **reflect** at each step, just like this student.

## How Self RAG works

```
Question
   |
   v
[1] Do I need to retrieve? --- No ---> Answer directly
   | Yes
   v
[2] Retrieve documents
   |
   v
[3] Are they relevant?  --- No ---> Search again / ignore them
   | Yes
   v
[4] Generate an answer
   |
   v
[5] Does the answer match the documents? --- No ---> Fix the answer
   | Yes
   v
Final answer
```

## The key idea: self-grading

Self RAG adds **self-check steps** (called reflection):

- **Relevance check** → "Is this document useful?"
- **Support check** → "Is my answer supported by the document?"
- **Usefulness check** → "Is this a good answer to the question?"

If any check fails, the AI **tries again** instead of giving a bad answer.

## A simple idea in pseudo-code

This is a *simplified* picture of the self-checking logic (not a full app).

````python
def self_rag(question, documents):
    # Step 1: is any document relevant?
    relevant = [d for d in documents if is_relevant(d, question)]

    if not relevant:
        return "No good information found. Need to search again."

    # Step 2: build an answer from relevant docs
    answer = make_answer(relevant)

    # Step 3: self-check -- is the answer supported?
    if not is_supported(answer, relevant):
        return "Answer not supported. Try again."

    return answer


# Simple helper functions (just for demonstration)
def is_relevant(doc, question):
    return any(word in doc.lower() for word in question.lower().split())

def make_answer(docs):
    return "Based on the data: " + docs[0]

def is_supported(answer, docs):
    return True   # in a real app, the AI would judge this


# Try it
docs = ["Python is a programming language.", "Cats sleep a lot."]
print(self_rag("What is python?", docs))
````

## Key points to remember

- Self RAG = the AI **checks its own work**.
- It decides **when** to search (not always needed).
- It grades documents for **relevance**.
- It checks if the answer is **supported** by the data.
- If a check fails, it **retries** — giving more reliable answers.
