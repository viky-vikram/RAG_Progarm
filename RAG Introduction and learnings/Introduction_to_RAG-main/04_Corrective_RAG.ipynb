# 4. Corrective RAG (CRAG)

**Corrective RAG** (CRAG) is a type of RAG that **checks the quality of the retrieved data** and **corrects itself** if the data is bad.

If the information it finds is **wrong or not enough**, it does **not** give a poor answer. Instead, it **searches again** — often from another source like the **web**.

## Simple real-life example

Imagine you ask a friend a question:

- Your friend looks in a book but the page is **not helpful**.
- Instead of guessing, your friend says: "This isn't enough — let me **Google it**."
- Then gives you a correct answer from the new, better source.

That is exactly what Corrective RAG does.

## How Corrective RAG works

```
Question -> Retrieve documents
                 |
                 v
         Grade the documents
          /       |        \
     CORRECT    AMBIGUOUS    WRONG
        |          |          |
     use them   use them    throw away
        |       + web search    + web search
        \_________ | _________/
                   v
             Generate answer
```

## The key idea: grade, then correct

CRAG adds a **grader** that labels the retrieved data as:

- ✅ **Correct** → the data is good, use it.
- ⚠️ **Ambiguous** → not sure, so use it **plus** extra web search.
- ❌ **Incorrect** → the data is bad, **throw it away** and search the web.

This "correction" step is what gives CRAG its name.

## A simple idea in pseudo-code

````python
def corrective_rag(question, documents):
    grade = grade_documents(question, documents)

    if grade == "correct":
        data = documents
    elif grade == "ambiguous":
        data = documents + web_search(question)
    else:  # incorrect
        data = web_search(question)   # ignore bad docs, use web

    return "Answer using: " + str(data)


# Simple helper functions (just for demonstration)
def grade_documents(question, documents):
    # pretend the docs are not good enough
    return "incorrect"

def web_search(question):
    return ["(fresh result from the web about: " + question + ")"]


print(corrective_rag("Latest news about AI?", ["old unrelated text"]))
````

## Self RAG vs Corrective RAG

- **Self RAG** → the AI checks **its own answer** and reasoning.
- **Corrective RAG** → the AI checks the **retrieved data** and re-searches (often the web) if it is bad.

They both add a "check" step, but they check **different things**.

## Key points to remember

- Corrective RAG = **check the data quality**, then fix it.
- A **grader** labels data as correct, ambiguous, or incorrect.
- If data is bad, it **searches again** (often the web).
- This avoids answers based on **wrong information**.
- It mainly checks the **retrieved data** (not the answer itself).
