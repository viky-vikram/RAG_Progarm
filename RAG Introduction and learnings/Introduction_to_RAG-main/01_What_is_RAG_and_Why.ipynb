# 1. What is RAG and Why it is Used

**RAG** stands for **Retrieval-Augmented Generation**.

It is a method where an AI model **first searches for relevant information**, and then **uses that information to answer** your question.

In short:
> **Retrieval** = find the right data.
> **Augmented** = add that data to the question.
> **Generation** = the AI writes the answer.

## The problem RAG solves

A normal AI model (like ChatGPT) has two big problems:

1. **It does not know your private data** (your notes, company files, PDFs).
2. **It can "hallucinate"** — make up answers that sound correct but are wrong.

RAG fixes both by giving the AI the **real, correct information** before it answers.

## Simple real-life example

Imagine a student in an exam:

- **Without RAG (closed book):** the student answers only from memory — may forget or guess wrongly.
- **With RAG (open book):** the student first looks up the correct page, then writes the answer.

RAG is like giving the AI an **open book** to read from before answering.

## How RAG works (the basic flow)

```
User asks a question
        |
        v
1. RETRIEVE  -> search your data for relevant text
        |
        v
2. AUGMENT   -> add that text to the question
        |
        v
3. GENERATE  -> AI reads it and writes the final answer
```

## Why RAG is used (benefits)

- ✅ **Accurate answers** — based on real data, not guesses.
- ✅ **Works with your own data** — PDFs, notes, company documents.
- ✅ **Always up to date** — just update the data, no need to retrain the AI.
- ✅ **Reduces hallucination** — fewer made-up answers.
- ✅ **Cheaper** — no need to retrain huge models.

## Key points to remember

- RAG = **Retrieval-Augmented Generation**.
- It **searches for data first**, then lets the AI answer using it.
- It is like giving the AI an **open book**.
- The 3 steps: **Retrieve ➜ Augment ➜ Generate**.
- It gives accurate answers and reduces hallucination.
